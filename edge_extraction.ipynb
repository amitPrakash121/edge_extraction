This is based on "Fast and robust edge extraction in unorganized point clouds" by Dena Bezazian, Josep R. Casas and Javier Ruiz-Hidalgo published in 2015.

The point cloud is in txt format, comma separated with values x, y, z, nx, ny, nz. The normals will not be required in this. The file has 10029 points.

# Part 1: Save the point cloud in a dictionary with serial no starting from 1 as key and value as a list containing a list of x, y, z

```python
file_name = "Part1_STL_10_pow_4_ASCII.txt"
```

```python
with open(file_name) as file_object:
    lines = file_object.readlines()
```

```python
data = {}
```

```python
i = 1
for line in lines:
    items = line.split(",")
    x, y, z = float(items[0]), float(items[1]), float(items[2])
    list_x_y_z = [x, y, z]
    data[i] = list_x_y_z
    i+=1
```

# Part 2: Calculate nearest neighbours (taking 8 nearest neighbours) based on euclidean distance and add keys to dictionary

```python
import math
```

```python
def nearest_neighbour(key):
    key_list = []
    dict1 = {}
    for i in range(1, len(data)+1):
        if i == key:
            continue
            
        euclidean_distance = math.sqrt((data[key][0] - data[i][0])**2 + (data[key][1] - data[i][1])**2 + (data[key][2] - data[i][2])**2)
        dict1[i] = euclidean_distance
        
    # Sort dictionary by value in ascending order
    for i in range(8):
        for key1 in dict1.keys():
            min_key = key1
            break
        for key1 in dict1.keys():
            if dict1[key1] < dict1[min_key]:
                min_key = key1
        key_list.append(min_key)
        del dict1[min_key]
    return key_list
```

```python
for i in range(1, len(data)+1):
    nearest_neighbour_list = nearest_neighbour(i)
    data[i].append(nearest_neighbour_list)    
```

# Part 3: Calculate surface variation sigma_p from eigen values of the covariance matrix of each point and append it to dictionary. For flat surface sigma_p tends to zero

```python
import numpy as np
```

```python
def sigma_p(key):
    covariance_matrix = np.zeros((3,3))
    mean_x, mean_y, mean_z = 0, 0, 0
    for neighbour in data[key][3]:
        mean_x += data[neighbour][0]
        mean_y += data[neighbour][1]
        mean_z += data[neighbour][2]
        
    mean_x = mean_x/8
    mean_y = mean_y/8
    mean_z = mean_z/8
    cov_x_x = 0
    cov_y_y = 0
    cov_z_z = 0
    cov_x_y = 0
    cov_x_z = 0
    cov_y_z = 0
    
    for neighbour in data[key][3]:
            cov_x_x += (data[neighbour][0] - mean_x)**2
            cov_y_y += (data[neighbour][1] - mean_y)**2
            cov_z_z += (data[neighbour][2] - mean_z)**2
            cov_x_y += (data[neighbour][0] - mean_x)*(data[neighbour][1] - mean_y)
            cov_x_z += (data[neighbour][0] - mean_x)*(data[neighbour][2] - mean_z)
            cov_y_z += (data[neighbour][1] - mean_y)*(data[neighbour][2] - mean_z)
            
    covariance_matrix[0,0] = cov_x_x/7
    covariance_matrix[1,1] = cov_y_y/7
    covariance_matrix[2,2] = cov_z_z/7
    covariance_matrix[0,1] = cov_x_y/7
    covariance_matrix[1,0] = cov_x_y/7
    covariance_matrix[0,2] = cov_x_z/7
    covariance_matrix[2,0] = cov_x_z/7
    covariance_matrix[1,2] = cov_y_z/7
    covariance_matrix[2,1] = cov_y_z/7
    
    eigenValue, eigenVector = np.linalg.eig(covariance_matrix)
    
    sigmaP = min(eigenValue)/sum(eigenValue)
    
    return sigmaP
```

```python
for i in range(1, len(data)+1):
    sigma = sigma_p(i)
    data[i].append(sigma)
```
